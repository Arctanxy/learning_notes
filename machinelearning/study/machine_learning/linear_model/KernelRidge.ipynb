# Kernel Ridge ——核岭回归

核岭回归就是在岭回归中加入核函数，将非线性的数据投射到核空间，再进行线性拟合。

岭回归的目标函数是：

$$\sum_{i=1}^{n}||y-X\beta|| + \lambda \beta^T \beta$$

但是原始数据并不总是那么理想，直接使用岭回归效果可能不太好，这时可以将数据映射到一个核空间，使数据满足线性规律。

变换可以理解为对X进行非线性变换，使新的数据的维度上升（增加了一些与X相关的维度），映射到高维数据空间之后就可以使用线性拟合了。假设$\Phi_i = \Phi(x_i)$，则岭回归的目标函数变成了:

$$\sum_{i=1}^{n}||y-\Phi\beta||^2 + \lambda \beta^T\beta$$

最优化结果：

$$\beta = \Phi^T(\Phi\Phi^T + \lambda I_n)^{-1}y$$

具体的推导过程可以参看：http://www.cnblogs.com/wacc/p/3751632.html

没有接触过核函数的话，可以以多项式回归来理解核函数，假设$\Phi(x) = ax^3 + bx^2 + cx + d$，这个核函数就相当于在原数据X的基础上添加了两个维度：$x^3$和$x^2$。

以下面的数据为例，来看看映射到高维空间之后线性回归效果的改善：

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
x = np.linspace(1,10,100)
y = [np.sin(n) + 2 * np.random.random() for n in x]
y = np.array(y)
plt.scatter(x,y)
```

** 添加维度 **

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
x2 = x ** 2
x3 = x ** 3
x_new = np.concatenate((x,x2,x3),axis=1)
# x = np.transpose(x)
x_new.shape,y.shape
```

在sklearn中提供了核岭回归模型，可以很方便地调用：

```python
from sklearn.kernel_ridge import KernelRidge
model = KernelRidge(kernel='rbf')
model.fit(x,y)
y_pred = model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
```

拟合效果比我们自己用的三次函数好得多~

这里用到到核函数为RBF(径向基函数），RBF函数能够将低维数据映射到高维空间，从而使线性不可分的数据变得线性可分，常用与SVM（支持向量机）中，在线性回归中同样使用。

在sklearn提供了多个可选的核函数，分别是：rbf，linear，laplacian, polynomial 和 sigmoid ，一般来说大多数问题使用rbf都能获得较好的效果，下面我们就对其他核函数进行一一试验：

```python
model = KernelRidge(kernel='linear')
model.fit(x,y)
y_pred = model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
```

```python
model = KernelRidge(kernel='laplacian')
model.fit(x,y)
y_pred = model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
```

```python
model = KernelRidge(kernel='polynomial')
model.fit(x,y)
y_pred = model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
```

```python
model = KernelRidge(kernel='sigmoid')
model.fit(x,y)
y_pred = model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
```
