## 2. 技能要求

```python
import jieba
des = '--'.join(df['DESCRIPTION'].values)
des = des.lower()
des = list(jieba.cut(des))
requires = dict(Counter(des))
# requires = sorted(requires.items(),key=lambda x:x[1],reverse=True)
requires
```

```python
#### 编程语言
'''
Python、Java、C++、C、R
'''
lang = {}
lang['python'] = requires['python']
lang['java'] = requires['java']
lang['c++'] = requires['c++']
lang['c'] = requires['c']
lang['r'] = requires['r']
lang['matlab'] = requires['matlab']
lang['scala'] = requires['scala']
lang
```

```python
# 显而易见，Python是机器学习首选语言，而Java和C++旗鼓相当，可以任选其一，这里我选择了C++
```

```python
#### 框架
frame = {}
frame['tensorflow'] = requires['tensorflow'] + requires['tf']
frame['pytorch'] = requires['torch'] + requires['pytorch']
frame['theano'] = requires['theano']
frame['caffe'] = requires['caffe']
frame['weka'] = requires['weka']
frame['dl4j'] = requires['dl4j']
frame['mxnet'] = requires['mxnet']
frame
```

TensorFlow雄踞榜首，但是TensorFlow版本之间的变动太大，不适合我这种业余时间学习的人看，所以框架我选择了最简单易用的Pytorch，如果工作需要用到其他的再转

```python
#### 其他高频词汇
high = {}
for k,v in requires.items():
    if v >= 100:
        high[k] = v
high
```

```python
import re
df['low_salary'] = df['salary'].apply(lambda x:x.split('-')[0] if '-' in x else x)
df['high_salary'] = df['salary'].apply(lambda x:x.split('-')[1] if '-' in x else x)
df['low_salary'] = df['low_salary'].apply(lambda x: float(re.search(r'\d+',x).group()))
df['high_salary'] = df['high_salary'].apply(lambda x: float(re.search(r'\d+',x).group()))
df['low_salary'].hist(bins=20,alpha=0.4),df['high_salary'].hist(bins=20,alpha=0.4)
```
