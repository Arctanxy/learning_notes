```python
def get_prices(num):
    data24 = data2[data2['总楼层'] == num]
    f_p = {}
    for name,group in data24.groupby('所在楼层'):
        if name != 1:
            f_p[name] = group['单价'].mean()
    plt.plot(f_p.keys(),f_p.values())
```

```python
data2 = data2[(data2['所在楼层'] != 1)&(data2['所在楼层'] != data2['总楼层'])]
data2['ratio'] = data2['所在楼层']/data2['总楼层'] # 楼层位置
pivot_price = {}
for name,group in data2.groupby('小区名称'):
    pivot_price[name] = group['单价'].mean()
data2['pivot'] = data2['小区名称'].apply(lambda x: pivot_price[x])
data2['ratio_price'] = data2['单价']/data2['pivot']
plt.figure(figsize=(10,6))
plt.scatter(data2['ratio'],data2['ratio_price'])
```

```python
data2 = data2[data2['ratio_price'] <= 1.2]
x = data2['ratio']
y = data2['ratio_price']
plt.figure(figsize=(10,6))
plt.scatter(x,y)
```

```python
x.reshape(-1,1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np 
pf = PolynomialFeatures(degree=2)
x = x.reshape(-1,1)
x = pf.fit_transform(x)
x
```

```python
from sklearn.linear_model import RidgeCV
model = RidgeCV(alphas=[1e-4,1e-5,1e-3,1e-2,1e-1,1e0])
model.fit(x,y)
yp = model.predict(x)
loss = np.abs(yp-y)/y
np.mean(loss)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x[:,1],y)
plt.scatter(x[:,1],yp)
```

```python
model.intercept_,model.coef_
```

系数
$$K = -0.1779188*(\frac{n}{N})^2 + 0.20702933*\frac{n}{N} + 0.949938$$

评估价格
$$P = P_{jizhun} * K$$

其中$N$是总楼层，$n$是存量房所在楼层

```python
coef = np.array([0.,0.21,-0.18]).reshape(-1,1)
x.shape
```

```python
yp2 = np.dot(x,coef) + 0.95
plt.figure(figsize=(10,5))
plt.scatter(x[:,1],y)
plt.scatter(x[:,1],yp2)
```
