```python
import pandas as pd 
import numpy as np 
from scipy.optimize import leastsq
```

```python
data = pd.read_excel("D:/修改公式/未修正小区价格.xlsx")[4300:]
data.head(100)
```

# 拿一个小区出来做实验

```python
df = data[(data['小区名称']=="众德祥福园")&(data['类型']=='中高层住宅')]

def evaluation(df,p):
    df['ratio'] = df['所在楼层']/data['总楼层']
    # a,b,c = p
    # df['xishu'] = df['ratio'].apply(lambda x:a*pow(x,2) + b * x + c)
    df['xishu'] = p
    
    df['比准价格'] = df['单价']/df['xishu']
    df['基准价格'] = df['比准价格'].mean()
    df['评估价格'] = df['基准价格'] * df['xishu']
    return df['评估价格'].values
def residual(p,y,x):
    return y - evaluation(x,p)
y = df['单价'].values
x = df
p0 = [0.1 for i in range(df.shape[0])]
plsq = leastsq(residual,p0,args = (y,x))
yp = evaluation(df,plsq[0])
np.mean(residual(plsq[0],y,x))
```
