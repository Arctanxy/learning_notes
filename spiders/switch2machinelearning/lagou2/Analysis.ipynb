## 1. 职位标签

```python
from functools import reduce
from collections import Counter
df['positionLables'] = df['positionLables'].map(eval)
labels = reduce(lambda x,y:x+y, df['positionLables'].values) # 从二维列表变一维列表
Counter(labels)
```

```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt 
from scipy.misc import imread
%matplotlib inline
positionlables = '.'.join(labels)
bg = imread('blank.bmp')
wc = WordCloud(
        #设置字体，不指定就会出现乱码，注意字体路径
        font_path="simkai.ttf",
        #font_path=path.join(d,'simsun.ttc'),
        #设置背景色
        background_color='white',
        #词云形状
        mask=bg,
        #允许最大词汇
        max_words=2000,
        #最大号字体
        max_font_size=160
    )

wc.generate(positionlables) # 产生词云
wc.to_file("positionLabel.png")
plt.imshow(wc,shape=(300,300))
```

```python
import re
df['low_salary'] = df['salary'].apply(lambda x:x.split('-')[0] if '-' in x else x)
df['high_salary'] = df['salary'].apply(lambda x:x.split('-')[1] if '-' in x else x)
df['low_salary'] = df['low_salary'].apply(lambda x: float(re.search(r'\d+',x).group()))
df['high_salary'] = df['high_salary'].apply(lambda x: float(re.search(r'\d+',x).group()))
df['low_salary'].hist(bins=20,alpha=0.4),df['high_salary'].hist(bins=20,alpha=0.4)
```

```python
for city in ['北京','上海','深圳','杭州','成都','广州','武汉','南京']:
    low_salary = df[df['city'] == city]['low_salary'].mean()
    high_salary = df[df['city'] == city]['high_salary'].mean()
    print(city,low_salary,high_salary)
```

```python
house_price = {
    '北京':54909,'上海':50550,'深圳':49889,'杭州':20504,'成都':13114,
    '广州':30016,'武汉':16589,'南京':25930
}
for city in ['北京','上海','深圳','杭州','成都','广州','武汉','南京']:
    low_salary = df[df['city'] == city]['low_salary'].mean()*14000/house_price[city]
    high_salary = df[df['city'] == city]['high_salary'].mean()*14000/house_price[city]
    print(city,low_salary,high_salary)
```
